# Visualizações com a biblioteca *Seaborn*

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
df = pd.read_csv('data/imigrantes_canada.csv')
df.set_index('País', inplace=True)
df
```

```python
top_10 = df.sort_values('Total', ascending=False).head(10)
top_10
```

```python
sns.barplot(data=top_10,
            x=top_10.index,
            y='Total')

```

```python
ax = sns.barplot(data=top_10,
                 x='Total',
                 y=top_10.index,
                 orient='h')

ax.set(title='Países com maior número de imigrantes para o Canadá\n(1980 - 2013)',
       xlabel='Número de imigrantes',
       ylabel='')


plt.show()
```

```python
def gerar_barplot_h(palette=None):
    fig, ax = plt.subplots(figsize=(10,5))

    ax = sns.barplot(data=top_10,
                    x='Total',
                    y=top_10.index,
                    hue=top_10.index,
                    orient='h',
                    palette=palette # 'None' retornará a paleta de cores padrão do Seaborn
                    )

    ax.set_title('Países com maior número de imigrantes para o Canadá\n(1980 - 2013)', fontsize= 14)
    ax.set_xlabel('Número de imigrantes', fontsize=10)
    ax.set_ylabel('')

plt.show() 
```

```python
gerar_barplot_h()
plt.show()
```

```python
sns.set_theme(style='dark')

gerar_barplot_h()
plt.show()
```

```python
sns.set_theme(style='ticks')
gerar_barplot_h()
sns.despine()
plt.show()
```
