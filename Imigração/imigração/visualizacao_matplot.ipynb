# Visualizações com a biblioteca *Matplotlib*

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Lendo o  dataset

```python
df = pd.read_csv('data/imigrantes_canada.csv')
df
```

#### Fixando 'País' como index

```python
df.set_index('País', inplace=True)
df
```

```python
anos = df.columns[2:-1].to_list()
anos
```

```python
brasil = df.loc['Brasil', anos]
brasil
```

```python
brasil_dict = {
    'ano': anos,
    'imigrantes':brasil.values.tolist()
}
```

```python
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

## Plotando um gráfico

```python
plt.plot(dados_brasil['ano'],
         dados_brasil['imigrantes'])
plt.xticks(dados_brasil.loc[::5,'ano'])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.show()
```

#### Comparando com outros países da América do Sul

```python
america_sul = df['Região'] == 'América do Sul'
df.loc[america_sul,anos].sum(axis=1).sort_values(ascending=False)
```

```python
dados_argentina = df.loc['Argentina',anos]
```

#### Gráfico com mais de um dado

```python
plt.plot(dados_argentina.index, dados_argentina.values, label='Argentina')
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'], label='Brasil')
plt.xticks(anos[::5])
plt.title('Imigrações Brasil e Argentina para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.legend()
plt.show()
```

### Explorando subplots

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'],
        lw=2,
        marker='o',
        color='mediumpurple')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Imigração do Brasil para o Canadá\n(1980-2013)',
             fontsize=18)
ax.set_xlabel('Ano',
              fontsize=12)
ax.xaxis.set_tick_params(labelsize=8)
ax.set_ylabel('Número de imigrantes',
              fontsize=12)
ax.yaxis.set_tick_params(labelsize=8)
plt.grid(linestyle='--')
plt.show()
```

#### Múltiplos gráficos

```python
fig, axs = plt.subplots(nrows=1,
                        ncols=2, figsize=(15,5))
# Primeiro gráfico
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n(1980-2013)')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

# Segundo gráfico
axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da imigração do Brasil para o Canadá\n(1980-2013)')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')
axs[1].grid()

plt.show()
```

```python
fig, axs = plt.subplots(nrows=2,
                        ncols=2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de imigrantes')
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_ylim(0, 7000)

plt.show()
```

```python
america_sul_sorted = df.loc[america_sul, 'Total'].sort_values(ascending=True)
america_sul_sorted
```

#### Customizando o gráfico

```python
# colors = ['indianred', 'darkorange', 'gold', 'olive', 'forestgreen', 'lime',
#           'aqua', 'mediumturquoise', 'deepskyblue', 'royalblue', 'mediumslateblue', 'mediumorchid']

colors = ['green' if pais == 'Brasil' else 'silver' for pais in america_sul_sorted.index]

fig, ax = plt.subplots(figsize= (12,6))
ax.barh(america_sul_sorted.index, america_sul_sorted,
       color=colors)
ax.set_title('Imigração da América do Sul para o Canadá\n(1980 - 2013)',
             fontsize=14)
ax.set_xlabel('Número de imigrantes',
              fontsize=12)
plt.show()
```

```python
# colors = ['indianred', 'darkorange', 'gold', 'olive', 'forestgreen', 'lime',
#           'aqua', 'mediumturquoise', 'deepskyblue', 'royalblue', 'mediumslateblue', 'mediumorchid']

colors = ['green' if pais == 'Brasil' else 'silver' for pais in america_sul_sorted.index]

fig, ax = plt.subplots(figsize= (12,6))
ax.barh(america_sul_sorted.index, america_sul_sorted,
       color=colors)
ax.set_title('Imigração da América do Sul para o Canadá\n(1980 - 2013)',
             fontsize=14)
ax.set_xlabel('Número de imigrantes',
              fontsize=12)
ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

for i, v in enumerate(america_sul_sorted):
    ax.text(v+50, i, str(v), ha='left', va='center')  # "v+50" é o distanciamento do número em relação à barra

fig.savefig('imigracao_america_do_sul.png', transparent=False,dpi=300, bbox_inches='tight')
plt.show()
```
